### Cleanning Data

Before that we need to add some columns to get the duration of rides, hour of the day, week, and month of rides. This can be done with the help of started_at and ended_at columns.We need to make sure that all the units are constant in the data:

```r
duration <- as.double(difftime(AllTrips_2021$ended_at, AllTrips_2021$started_at, units = "m"))
#To keep the data constant, we will use same unit
```

The cleaning process is important to take out the unnecessary columns which are not needed for the specific analysis. We can do that with the function and add columns by adding a pipe:

```r
#removing columns and adding columns
AllTrips_2021_2 <- AllTrips_2021 %>%
  select(-c(ride_id, started_at, ended_at, start_station_id, end_station_id, start_lat, start_lng, end_lat, end_lng)) %>%  add_column(mins_duration = duration, hour_trip = format(AllTrips_2021$started_at, format = "%H:%M"), 
               weekday = format(as.Date(AllTrips_2021$started_at), '%a'), day = format(as.Date(AllTrips_2021$started_at), '%d'), month = format(as.Date(AllTrips_2021$started_at),'%b_%y'))
```

Let check the columns added and removed.

```r
head(AllTrips_2021_2)
```

Despite cleaning up our data, we still need to check inaccurate information. To do it, we must be aware of the potential dangers of faulty data. Since it is impossible to have a negative trip length, we must clean the data for a more accurate analysis. To do this, we must count the number of rows with trip duration less than 0.

```r
nrow(subset(AllTrips_2021_2, mins_duration < 0))
```

We can see that there were 147 rows that has less than zero trip duration.

```r
#Cleaning negative values
AllTrips_2021_2x <- AllTrips_2021_2[!(AllTrips_2021_2$mins_duration < 0),]
```

# Analysis - Phase 04

The data set is now clean and is ready to be analyzed. For that we would need to keep in mind some information, which we can get with the summary function.

```r
summary(AllTrips_2021_2x$mins_duration)
```

```r
AllTrips_2021_2x %>%
    group_by(member_casual) %>%
    summarise(min_trip_duration = min(mins_duration),max_trip_duration = max(mins_duration),
              median_trip_duration = median(mins_duration), mean_trip_duration = mean(mins_duration), sd_trip_duration = sd(mins_duration))
```

We can see how the mean trip duration of casual riders is more than the mean duration of all the trips. casual mean trip is 32.00221, while members have a mean of 13.63355. This gives us information of how casual riders would ride more than the members on average. But this can be due to more number of rides by members than casual.

Our objective of research and analysis has been the behavior of casual and member. We need to keep in mind the different variables that would be useful for the behavioral analysis. We would need:

> 1. Hourly analysis
> 2. Weekly analysis
> 3. Monthly analysis

So lets break them down:

```r
#Order of weekday and month variable
AllTrips_2021_2x$weekday <- ordered(AllTrips_2021_2x$weekday, levels=c("Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"))
AllTrips_2021_2x$month <- ordered(AllTrips_2021_2x$month, levels=c("Jan_22", "Feb_22", "Mar_22", "Apr_22", "May_22", "Jun_22", "Jul_22"))
AllTrips_2021_2x %>% 
  group_by(member_casual, weekday) %>%  
  summarise(number_of_rides = n(),average_duration_mins = mean(mins_duration)) %>% 
  arrange(member_casual, desc(number_of_rides))
```

Lets start with our **Hourly Analysis** and create a visualization with ggplot2:

```r
daily_analysis <- AllTrips_2021_2x %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_trips = n(),average_duration_mins = mean(mins_duration)) %>% 
  arrange(member_casual, desc(number_of_trips))

daily_analysis_plot <- daily_analysis %>% 
  ggplot() +
  labs("trips by customer type vs weekday") +
  theme(axis.text.x = element_text(angle = 20)) +
  geom_col(aes(x = weekday, y = number_of_trips, fill = member_casual), width = 1, position = position_dodge(width = 0.5)) + 
scale_fill_manual(values = c("member" = "blue", "casual" = "green"))

print(daily_analysis_plot)
print(daily_analysis)
```

**Key Observation**
> 1. Members usually have more number of trips on weekdays. 
   
   This can be due to commuting to work.
   
> 2. Casual users number of trips are more in the weekends.  
   
   The reason behind this can be that they ride it when they have free time or on weekends casually.

**Monthly Analysis:**

```r
AllTrips_2021_2x%>% 
  group_by(member_casual, month) %>%  
  summarise(number_of_trips = n(),`average_duration_mins` = mean(mins_duration)) %>% 
  arrange(member_casual,desc(number_of_trips))
```

lets check the **Monthly Analysis** and create a visualization with ggplot2 to make it more clear:

```r
monthly_analysis <- AllTrips_2021_2x %>% 
  group_by(member_casual, month) %>% 
  summarise(number_of_trips = n(),average_duration_mins = mean(mins_duration)) %>% 
  arrange(member_casual, desc(number_of_trips))

monthly_analysis_plot <- monthly_analysis %>% 
  ggplot() +
  labs("trips by customer type vs monthly") +
  theme(axis.text.x = element_text(angle = 20)) +
  geom_col(aes(x = month, y = number_of_trips, fill = member_casual), width = 1, position = position_dodge(width = 0.5)) + 
scale_fill_manual(values = c("member" = "blue", "casual" = "green"))

print(monthly_analysis_plot)
```

**Key Observations:**
1. Members have had more trips than casual riders in all 12 months of 2021.
2. Members usually use their rides for commute to nearby places which is why they have more number of rides than the casual users.
3. Casual and members have increased number of rides in the summer. This may be because of holidays and the weather, where riding bicycle is much more fun, while members can be concluded as commuters all around the year.

# Act - Phase 06

**Recommendations:**

As our priority objective is to engage the casual users and entice them to pay for the annual membership, here are some casual user focused recommendations:
1. Promotions on weekdays to engage them to use the bikes on weekdays too. This will change their user behavior and would assist them to buy the membership.

2. Casual riders may be young people who are doing it for leisure. We can provide student discounts and promotions that would engage with the college population. Making it easier and cheaper for students would help to get their interest and could lead to more memberships sold.

3. pricing models also help. Providing 3-4 different price brackets for the service might give them more detail on how this annual membership can be practical for them.

4. Partnering with sports club or gyms might also engage more people to buy the membership. The app could have a fitness tracker that would readily count your calories and give other data to keep the members engaged. There can be milestones and rewards according to them.

5. Weather apps can be synced in the app in order to give the riders more detail on what to expect. 

6. Some stations could be labeled as reward stations and these can be circled every week to different areas and locations. Cyclistic could partner with some food chains giving them some discounts if a user gets his bike or parks his bike on those particular station.

**Conclusions & Recommendations:**

Our main goal is to attract casual users and entice them to pay an annual fee, so here are our recommendations for casual users.

2. Casual drivers may be young people who drive for leisure purposes. Offer student discounts and promotions for college students. Making it easier and cheaper for students can keep interest and potentially lead to more membership sales.

3. Pricing models also help. By offering 3-4 different price points for the service, you can get more information on how this annual membership can come in handy.
